```python
#Clean the wiki movie list by making a copy and cleaning the copy. Source will still untouched
#Consolidate and the columns that have alternative titles into one column so there will be less columns using a loop
#check if the current key exists in the movie object. If it is remove the key-value pair and add to the alt title dict
#After looping through every key, add the alt titles dict to the movie object

def clean_movie(movie):
    movie = dict(movie) #create a non-destructive copy
    alt_titles = {}
    for key in ['Also known as','Arabic','Cantonese','Chinese','French',
                'Hangul','Hebrew','Hepburn','Japanese','Literally',
                'Mandarin','McCune–Reischauer','Original title','Polish',
                'Revised Romanization','Romanized','Russian',
                'Simplified','Traditional','Yiddish']:
        if key in movie:
            alt_titles[key] = movie[key]
            movie.pop(key)
    if len(alt_titles) > 0:
        movie['alt_titles'] = alt_titles
        
        
    return movie
```

```python
#make a list of cleaned movies with a list comprehension

clean_movies = [clean_movie(movie) for movie in wiki_movies]
```

```python
# set the wiki_movies_df to be the dataframe created from the clean_movies

wiki_movies_df = pd.DataFrame(clean_movies)
sorted(wiki_movies_df.columns.tolist())
```
